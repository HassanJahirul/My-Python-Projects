### Data Cleanup

```python
df["Postal_Code"]=df["Postal_Code"].fillna(0)
```

```python
df["Postal_Code"]=df["Postal_Code"].astype(int)
df=df.convert_dtypes()
l=["Order_Date","Ship_Date"]
for i in l:
    df[i]=pd.to_datetime(df[i],format="%d-%m-%Y")
```
